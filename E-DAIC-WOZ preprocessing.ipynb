```python
import os
import pandas as pd
import pandas as pd

fold_path = 'D:\E-DAIC-WOZ\AUPOse'
file_path = os.listdir(fold_path)
n_list = []
train_list, dev_list, test_list = [], [], []
file_path.sort()
for i, file in enumerate(file_path):
  print(file[:3])
  if int(file[:3]) in train:
    train_list.append(pd.read_csv(os.path.join(fold_path, file), header=0, encoding_errors='ignore').iloc[:,4:].to_numpy())
  elif int(file[:3]) in dev:
    dev_list.append(pd.read_csv(os.path.join(fold_path, file), header=0, encoding_errors='ignore').iloc[:,4:].to_numpy())
  elif int(file[:3]) in test:
    test_list.append(pd.read_csv(os.path.join(fold_path, file), header=0, encoding_errors='ignore').iloc[:,4:].to_numpy())

with open(os.path.join(fold_path,'AUpose_train.pkl'),'wb') as f :
  pickle.dump(train_list, f)
with open(os.path.join(fold_path,'AUpose_dev.pkl'),'wb') as f :
  pickle.dump(dev_list, f)
with open(os.path.join(fold_path,'AUpose_test.pkl'),'wb') as f :
  pickle.dump(test_list, f)
```

```python
fold_path = 'E:\E-DAIC-WOZ\MFCC'
file_path = os.listdir(fold_path)
train_list, dev_list, test_list = [], [], []
file_path.sort()
for i, file in enumerate(file_path):

  print(file[:3])
  if int(file[:3]) in train:
    train_list.append(pd.read_csv(os.path.join(fold_path, file), header=0, encoding_errors='ignore', delimiter=';').iloc[:,4:].to_numpy())
  elif int(file[:3]) in dev:
    dev_list.append(pd.read_csv(os.path.join(fold_path, file), header=0, encoding_errors='ignore', delimiter=';').iloc[:,4:].to_numpy())
  elif int(file[:3]) in test:
    test_list.append(pd.read_csv(os.path.join(fold_path, file), header=0, encoding_errors='ignore', delimiter=';').iloc[:,4:].to_numpy())

with open(os.path.join(fold_path,'MFCC_train.pkl'),'wb') as f :
  pickle.dump(train_list, f)
with open(os.path.join(fold_path,'MFCC_dev.pkl'),'wb') as f :
  pickle.dump(dev_list, f)
with open(os.path.join(fold_path,'MFCC_test.pkl'),'wb') as f :
  pickle.dump(test_list, f)
```

```python
fold_path = 'D:\E-DAIC-WOZ\\text'
file_path = os.listdir(fold_path)
train_list, dev_list, test_list = [], [], []
file_path.sort()
for i, file in enumerate(file_path):
  print(file[:3])
  if int(file[:3]) in train:
    rows = pd.read_csv(os.path.join(fold_path, file), header=0, encoding_errors='ignore').iloc[:,2]
    for r in rows:
      r = r.split()
    train_list.append(r)

  elif int(file[:3]) in dev:
    rows = pd.read_csv(os.path.join(fold_path, file), header=0, encoding_errors='ignore').iloc[:,2]
    for r in rows:
      r = r.split()
    dev_list.append(r)
  
  elif int(file[:3]) in test:
    rows = pd.read_csv(os.path.join(fold_path, file), header=0, encoding_errors='ignore').iloc[:,2]
    for r in rows:
      r = r.split()
    test_list.append(r)

with open(os.path.join(fold_path,'text_train.pkl'),'wb') as f :
  pickle.dump(train_list, f)
with open(os.path.join(fold_path,'text_dev.pkl'),'wb') as f :
  pickle.dump(dev_list, f)
with open(os.path.join(fold_path,'text_test.pkl'),'wb') as f :
  pickle.dump(test_list, f)
```

```python
fold_path = 'D:\E-DAIC-WOZ\splits'
# file_path = os.listdir(fold_path)
def make_labels(mode = 'train'):
  file = mode + '_split.csv'
  data = pd.read_csv(os.path.join(fold_path, file), header=0, encoding_errors='ignore')
  labels = {k:v.to_list() for k, v in data.items() }
  
  with open(os.path.join(fold_path, mode+'.pkl'),'wb') as f :
    pickle.dump(labels, f)
  
  return labels

data = make_labels(mode='dev')
data

#   n_list
```
